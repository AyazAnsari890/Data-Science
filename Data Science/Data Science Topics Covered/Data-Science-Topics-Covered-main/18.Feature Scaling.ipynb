```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
# we standardized the data by using z-score formula where we need mean and standard deviation
import numpy as np
mean=np.mean(df['total_bill'])
std=np.std(df['total_bill'])
print(mean,std)
```

**Standardisation** This technique is to re-scale features value with the distribution value between 0 and 1 pasted image 0.png

```python
normalized_data=[]
for i in list(df['total_bill']):
    z_score=(i-mean)/std
    normalized_data.append(z_score)
```

```python
normalized_data
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(normalized_data)
```

```python
# We can perform the above activity through StandardScaler refer scikit-learn.org
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler() #we create an object of StandardScaler
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']]) #here it computes the mean and standard deviation as required for z-score.
```

```python
scaler.transform(df[['total_bill']]) #here it starts scaling down and applies z-score formula
```

```python
# Instead of writing above
scaler.fit_transform(df[['total_bill']])
```

```python
import pandas as pd
pd.DataFrame(scaler.fit_transform(df[['total_bill']]),columns=['total_bill'])
```

```python
# We can perform scaler operation on two columns also
df.head()
```

```python
import pandas as pd
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=['total_bill','tip'])
```

```python
# Example of using Feature Scaling on new dataset
scaler.transform([[13,4]])
```

## Normalization--Min Max Scaler

Normalization is a technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting differences in the ranges of values or losing information.

This is done to facilitate data analysis and modeling, and to reduce the impact of different scales on the accuracy of machine learning models. Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.

```python
df=sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max=MinMaxScaler()
```

```python
# Transforms data values between 0 to 1
min_max.fit_transform(df[['distance','fare','tip']])
```

```python
min_max.fit(df[['distance','fare','tip']])
```

```python
min_max.transform(df[['distance','fare','tip']])
```

```python
min_max.transform([[1.6,7.0,2.15]])
```
