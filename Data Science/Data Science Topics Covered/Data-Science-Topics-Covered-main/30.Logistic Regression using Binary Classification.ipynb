## Logistic Regression Implementation

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
# Gets the description of the dataset
print(dataset.DESCR)
```

```python
dataset.keys()
```

```python
print(dataset.feature_names)
```

```python
# Convert the dataset into dataframe
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
df
```

```python
print(dataset.target)
```

```python
# Add the output feature in our dataframe
df['target']=dataset.target
```

```python
df
```

```python
df.head()
```

```python
df['target'].unique()
```

Insight: The output feature has 3 values but for now we try to make it as binary classification so we take only two values for the o/p feature

```python
# shows the data which has target value 0 and 1 not 2
df[df['target']!=2]
```

```python
df_copy=df[df['target']!=2]
```

```python
df_copy['target'].unique()
```

```python
## Independent and dependent features
X=df_copy.iloc[:,:-1] #copies all the rows and column except last column which is output into X
y=df_copy.iloc[:,-1] #Copies the last column into y
```

```python
# Train the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classification=LogisticRegression(max_iter=200)
```

```python
classification
```

```python
classification.fit(X_train,y_train)
```

```python
# Predict the output for the test data
y_pred=classification.predict(X_test)
y_pred
```

## Confusion Metrics,Accuracy,Classification Report - Precision and Recall

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

Insight: Here the accuracy is showing 100 percent.

```python
#Classification_report shows the recall, precision, f1-score etc.
print(classification_report(y_test,y_pred))
```

## Cross Validation

```python
from sklearn.model_selection import KFold
```

```python
cv=KFold(n_splits=5)
```

```python
cv
```

```python
from sklearn.model_selection import cross_val_score
```

```python
## these are the scores all the cross validation
scores=cross_val_score(classification,X_train,y_train,scoring='accuracy',cv=cv)
```

```python
# Average of the accuracies
np.mean(scores)
```

## Lets see more complex data

```python
# make a prediction with a multinomial logistic regression model
# The make_classification function can be used to generate a random n-class classification problem.
# This initially creates clusters of points normally distributed (std=1).
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
# define dataset
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_classes=2, random_state=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
complex_class_model=LogisticRegression(max_iter=200)
```

```python
complex_class_model.fit(X_train,y_train)
```

```python
y_pred_complex=complex_class_model.predict(X_test)
```

## Confusion Metrics,Accuracy,Classification Report - Precision and Recall

```python
print(confusion_matrix(y_test,y_pred_complex))
print(accuracy_score(y_test,y_pred_complex))
print(classification_report(y_test,y_pred_complex))
```

```python
## Cross Validatiuon
cv=KFold(n_splits=5)
scores_complex=cross_val_score(complex_class_model,X_train,y_train,cv=cv)
scores_complex
```

```python
# Average of the accuracies
np.mean(scores_complex)
```
