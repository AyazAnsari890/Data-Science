```python
#SciPy is an open-source Python library which is used to solve scientific and mathematical problems.
#It is built on the NumPy extension and allows the user to manipulate and visualize data with a wide range of high-level commands
import scipy.stats as stat
import pylab
import numpy as np
```

```python
import seaborn as sns
```

```python
sns.load_dataset('iris')
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'sepal_width')
```

```python
plot_data(df,'petal_width')
```

```python
plot_data(df,'sepal_length')
```

```python
plot_data(df,'petal_length')
```

## **Chi Square with Python**

5IAt2rl2QzelSEz1SvJo_chi-square-equation2.jpg

```python
import scipy.stats as stat
import numpy as np
```

```python
## No of hours student study daily in a weekly basis
##monday,tuesday,Wednesday,thursday,Friday,Saturday,sunday
expected_data=[8,6,7,9,6,9,7]
observed_data=[7,8,6,9,9,6,7]
```

```python
sum(expected_data),sum(observed_data)
```

```python
## Chi square Goodness OF Fit
chisquare_test_statistics,p_value=stat.chisquare(observed_data,expected_data)
```

```python
print(chisquare_test_statistics),print(p_value)
```

```python
## find the critical value
significance=0.05
dof=len(observed_data)-1
critical_value=stat.chi2.ppf(1-significance,dof)
```

```python
critical_value
```

```python
if chisquare_test_statistics > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

##. **F Test with Python**

```python
#Evidence to say that the two population variances are not equal.
## Check we come to conclusion whether one worker is better than the other

worker1= [18, 19, 22, 25, 27, 28, 41, 45, 51, 55]
worker2= [14, 15, 15, 17, 18, 22, 25, 25, 27, 34]
```

```python
## Calculate the F test statistics
import numpy as np
f_test=np.var(worker1)/np.var(worker2)
```

```python
f_test
```

```python
## Degree of freedom
df1=len(worker1)-1
df2=len(worker2)-1
significane_value=0.05
```

```python
import scipy.stats as stat
```

```python
critical_value=stat.f.ppf(q=1-significane_value,dfn=df1,dfd=df2)
```

```python
if f_test>critical_value:
    print("Reject the Null Hypothesis")
else:
    print("We fail to reject the null hypothesis")
```
