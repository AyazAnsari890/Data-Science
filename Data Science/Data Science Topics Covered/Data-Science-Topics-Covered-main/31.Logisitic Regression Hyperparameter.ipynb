```python
# make a prediction logistic regression model
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
# define dataset
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_classes=2, random_state=1)
```

```python
# Train the model
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

## Logisitic Regression Hyperparameter Tuning

## Model Training Hyperparameter Tuning
## GridsearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'penalty':('l1', 'l2','elasticnet'),'C':[1,10,20,30]}
```

```python
classifier=LogisticRegression()
```

```python
clf=GridSearchCV(classifier,param_grid=parameters,cv=5)
```

```python
## Splitting of Training data to train and validation
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
classifier=LogisticRegression(C=1,penalty='l2')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
classifier.predict_proba(X_test)
```

## Confusion Metrics,Accuracy,Classification Report - Precision and Recall

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
